## Abstract
Stroke is the second leading cause of death worldwide, and a major cause of disability. Early detection and intervention are essential for reducing stroke mortality and morbidity. However, traditional stroke risk assessment methods are not very accurate, and there is a need for more sophisticated methods that can leverage the power of machine learning.

Early detection of strokes can be invaluable in the management and treatment of strokes. hence, the Random Forest Classifier was used to develop a machine learning model for the prediction and diagnosis of stroke from patient data. ## Introduction
Stroke diagnosis is a challenging task, even for experienced clinicians. This is due to the fact that stroke symptoms can vary widely from person to person, and some symptoms may be subtle or nonspecific. Additionally, stroke can be caused by a variety of underlying factors, making diagnosis even more difficult.

### Challenges associated with stroke diagnosis from patient data

Patient data such as gender, age, hypertension, heart disease, ever married, work type, Residence type, average glucose level, bmi and smoking status can be used to predict the risk of stroke. However, there are a number of challenges associated with using this data to diagnose stroke. One challenge is that the data may be incomplete or inaccurate. For example, patients may not be able to accurately recall their medical history, may not be aware of all of the symptoms they are experiencing or may not be truthful about their lifestyle habits.


Another challenge is that the relationship between stroke risk factors and stroke is complex. For example, some people with multiple risk factors for stroke may never have a stroke, while others with no known risk factors may experience a stroke. Finally, stroke diagnosis is often based on a combination of clinical symptoms and imaging tests. However, imaging tests may not be available immediately, and clinical symptoms can be nonspecific and mimic those of other conditions.


If the challenge of stroke diagnosis remains unsolved, it will have a number of implications for patients and healthcare systems. 
- For patients, a delayed or missed diagnosis of stroke can lead to serious complications, including death and disability. Even a small delay in treatment can have a significant impact on the patient's outcome. 
- For healthcare systems, a delayed or missed diagnosis of stroke can lead to increased costs and resource utilization. Patients who are misdiagnosed may require unnecessary tests and procedures, and they may also have longer hospital stays.


There are a number of ways to tackle the challenges associated with stroke diagnosis from patient data. One approach is to use machine learning techniques. Machine learning algorithms can be trained on large datasets of patient data to identify patterns that are associated with stroke. Once trained, these algorithms can be used to make predictions about whether or not a new patient is likely to have a stroke.


Another approach to tackling the challenges of stroke diagnosis is to develop new diagnostic tools. For example, researchers are developing new imaging techniques that can be used to detect stroke early. Additionally, researchers are developing new blood tests that can be used to identify biomarkers of stroke.


By addressing the challenges associated with stroke diagnosis, we can improve the outcomes for patients. Early diagnosis and treatment can lead to better quality of life and longer life expectancy.

### Preliminary blueprint of how I plan to tackle the challenge

- I propose to tackle the challenge of stroke diagnosis by developing machine learning models that can predict the risk of stroke based on patient data. The model will be trained on a large dataset of patient data, including data on stroke risk factors, clinical symptoms, and imaging tests.

Once the model is trained, it will be able to predict the risk of stroke for new patients based on their individual data. This information can then be used by doctors to make more informed decisions about stroke diagnosis and treatment.
- The performance of the models would be evaluated using the f1 score; an harmonic mean of the precision and recall scores.



I believe that this approach has the potential to improve the accuracy and timeliness of stroke diagnosis. This could lead to better outcomes for patients and reduce the burden on healthcare systems.



## Literature Review

Stroke is a serious medical condition that requires timely and accurate diagnosis and treatment. However, there are many challenges associated with stroke diagnosis, such as the lack of reliable and accessible biomarkers, the complexity and heterogeneity of stroke etiology and pathophysiology, and the variability and subjectivity of clinical symptoms and signs. These challenges may lead to delayed or missed diagnosis, inaccurate or inappropriate treatment, increased burden and cost of health care systems, and reduced quality of life and productivity of stroke survivors and caregivers. Therefore, there is a need for developing novel and effective methods for stroke diagnosis using patient data.


Patient data refers to the demographic, behavioral, and physiological information that can be collected from patients who are screened for stroke. Some examples of patient data are gender, age, hypertension, heart disease, ever married, work type, residence type, average glucose level, body mass index (BMI), and smoking status. These data can provide valuable insights into the risk factors and indicators of stroke, as well as the outcomes and prognosis of stroke patients. However, patient data are often incomplete, noisy, or inconsistent, which may limit their usefulness for stroke diagnosis.


Machine learning (ML) is a subfield of artificial intelligence (AI) that uses computerized algorithms to automatically learn from data and make predictions or decisions. ML can be applied to patient data to develop models that can diagnose stroke from various features and patterns. ML can also handle complex and high-dimensional data, as well as deal with uncertainty and variability in the data. ML can be divided into different types, such as supervised learning, unsupervised learning, reinforcement learning, and deep learning. Each type has its own advantages and disadvantages for stroke diagnosis.


In this literature review, we will focus on the application of supervised learning for stroke diagnosis using patient data. Supervised learning is a type of ML that uses labeled data, which means that the input and output variables are provided by human experts or annotators. The algorithm learns from the training data by making predictions and comparing them with the correct answers. The goal of supervised learning is to find a function that maps the input to the output accurately, so that it can make predictions on new data.


There are many studies that have used supervised learning for stroke diagnosis using patient data. We will review some of the most relevant and recent studies in this field, and discuss their methods, results, limitations, and implications. We will also compare them with our proposed project, which aims to use a data science approach to develop a machine learning model that can diagnose stroke from patient data.


Patil et al. (2022) used a synthetic dataset generated by the make_classification function from scikit-learnhttps://doi.org/10.3389/fmedt.2022.748949, which is a library that provides various tools and models for machine learning tasks. The dataset has 1,000 instances, 20 features, and two classes (stroke or not). The features are numerical and have 15 informative and 5 redundant features. The authors used an MLP classifier with two hidden layers of 16 units each to train and test their model for stroke diagnosis. They used accuracy score as their metric for evaluating their model's performance. They achieved an accuracy score of about 0.88 on the test set. 
The strengths of this study are:
- a synthetic dataset that can control the characteristics and distributions of the data was used. 
- MLP as a flexible and powerful algorithm that can handle nonlinear and complex data was used. 
- Class weight to balance the classes and reduce the bias towards the majority class.

The limitations of this study are: 
- A synthetic dataset that may not reflect the real-world patient data was used
- Accuracy score as the only metric, which may not be suitable for imbalanced data, as it does not account for the different costs or benefits of each type of error.
- No explanation or interpretation of the results or the features was provided

The implications of this study are: 
- It showed that patient data can be used to diagnose stroke with moderate accuracy using a complex algorithm. 
- It suggested that synthetic data can be a useful technique for testing and validating different algorithms and parameters for patient data. 
- It indicated that other metrics and techniques may be needed to improve the performance and understanding of the model.


Alegiani et al., 2020 aimed to assess the quality of stroke patient information applied in randomized controlled trials (RCTs) using machine learning (ML) methods. The authors collected 1,000 RCTs on stroke from PubMed and extracted 15 features from each trial, such as sample size, intervention type, outcome measure, and risk of bias. They used a supervised ML algorithm called random forest (RF) to classify the trials into high or low quality based on the Cochrane risk of bias tool. They also used an unsupervised ML algorithm called k-means clustering to group the trials into different clusters based on their features.

The strengths of this study are:
- A large and diverse dataset of RCTs on stroke from a reputable source was used. 
- RF as a robust and efficient algorithm that can handle high-dimensional and imbalanced data was used. 
- k-means clustering as a simple and intuitive algorithm that can reveal the patterns and trends in the data.

The limitations of this study are:
- Only 15 features were used to represent the RCTs, which may not capture all the relevant aspects of stroke patient information. 
- A binary classification for the quality of RCTs, which may not reflect the nuances and variations in the risk of bias was used. 
- No explanation or interpretation of the results or the features was provided.

The implications of this study are: 
- It showed that ML can be used to evaluate the quality of stroke patient information in RCTs.
- It suggested that there is room for improvement in the reporting and design of RCTs on stroke.
- It indicated that other methods and metrics may be needed to improve the validity and reliability of the ML models.

A Survey on Stroke Disease Classification and Prediction using Machine Learning Algorithms (Potdar et al., 2021) aimed to provide a comprehensive survey on the existing studies that have used ML algorithms for stroke disease classification and prediction. The authors reviewed 25 studies from various sources and summarized their objectives, methods, datasets, features, results, and limitations. They also compared and analyzed the performance of different ML algorithms for stroke diagnosis and prognosis.

The strengths of this study are:

- It discussed the advantages and disadvantages of each ML algorithm for stroke diagnosis and prognosis.

The limitations of this study are:
- It did not include any empirical or experimental evaluation of the ML algorithms or models.
- It did not provide any critical or constructive feedback on the existing studies or suggest any future directions or challenges.
- It did not use any citation or reference system to acknowledge the sources of the studies.

The implications of this study are:
- It showed that ML is a promising and popular technique for stroke disease classification and prediction.
- It provided a useful and informative overview of the current state-of-the-art in ML for stroke diagnosis and prognosis.
- It indicated that there is still room for improvement and innovation in ML for stroke disease classification and prediction.

F1 score is the harmonic mean of precision and recall, which are two measures of how well the model can identify the positive class (stroke) among all predictions. F1 score ranges from 0 to 1, where 1 is the best possible score and 0 is the worst. F1 score can be calculated as:


F1 = 2 * (precision * recall) / (precision + recall)

It's already established that the target has an imbalance class. Deleting about 16% of the minority class from the dataset will do us no good. Hence, We'd us imputation techniques to compute values for the missing values in the bmi - Body Mass Index.

Body Mass Index is a ratio of the body mass to the square height. Age is a good choice to consider in imputing missing entries for the bmi. 

Age and bmi may be checked for correlation

```python
data[['age', 'bmi', 'avg_glucose_level']].corr()
```

There is a weak positive linear relationship between age and bmi. As such median value of bmi for same age would be used to compute the missing value in the bmi

```python
df=data.copy()
```

```python
# Group the data by age
quartile_groups = data.groupby('age')

# Iterate through each group and calculate the median bmi as detailed earlier
quart_bmi_med = {quartle: group['bmi'].median() for quartle, group in quartile_groups}

# Iterate through each row in the dataset and fill in missing bmi values with the median for that quart_bmi_med
for index, row in data.iterrows():
    if pd.isna(row['bmi']):
        data.at[index, 'bmi'] = quart_bmi_med[row['age']]
```

### Dealing with Outliers

```python
numerical_variables = ['age', 'avg_glucose_level', 'bmi']

for i in numerical_variables:
    print(i)
    sns.boxplot(data=data[i])
    plt.show()
```

Presence of outliers is detected in bmi and average glucose level.

```python
#define a function to find outliers using IQR
def find_outliers(col):
    #calculate the first and third quartiles
    q1 = col.quantile(0.25)
    q3 = col.quantile(0.75)
    #calculate the interquartile range
    iqr = q3 - q1
    #calculate the lower and upper bounds
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    #return a boolean series indicating whether each value is an outlier
    return col[(col < lower_bound) | (col > upper_bound)]

```

```python
#apply the function to each column of the dataframe
agl_outliers = data[['avg_glucose_level']].apply(find_outliers)
agl_outliers
```

```python
a, b = agl_outliers.min().values[0], agl_outliers.max().values[0]
print(f'minimum value of outlier in average glucose level is {a}')
print()
print(f'maximum value of outlier in average glucose level is {b}')
```

```python
#apply the function to each column of the dataframe
bmi_outliers = data[['bmi']].apply(find_outliers)
bmi_outliers.min(), bmi_outliers.max()

c, d = bmi_outliers.min().values[0], bmi_outliers.max().values[0]
print(f'minimum value of outlier in BMI is {c}')
print()
print(f'maximum value of outlier in BMI is {d}')
```

Body mass index (BMI) is a person’s weight in kilograms divided by the square of height in meters. BMI is an inexpensive and easy screening method for weight category—underweight, healthy weight, overweight, and obesity.

BMI does not measure body fat directly, but BMI is moderately correlated with more direct measures of body fat 1,2,3. Furthermore, BMI appears to be as strongly correlated with various metabolic and disease outcome as are these more direct measures of body fatnes

For adults 20 years old and older, BMI is interpreted using standard weight status categories. These categories are the same for men and women of all body types and ages.

The standard weight status categories associated with BMI ranges for adults are shown in the following table.

BMI Weight Status

Below 18.5        Underweight

### Implementing Oversampling Techniques
The Random Over Sampler, Synthetic Minority Over-Sampling Technique (SMOTE) and Border line SMOTE would be used on the training set. These algorithms are selected because they are the most used over sampling algorithms.

```python
OverSamp_1 = RandomOverSampler(random_state=419)
OverSamp_2 = SMOTE(random_state=419, n_jobs=-1)
OverSamp_3 = BorderlineSMOTE(random_state=419, n_jobs=-1)

result_df = pd.DataFrame()

OverSamp_list = [OverSamp_1, OverSamp_2, OverSamp_3]
 

for model in [dtp, lrp, rfp, knp]:    
    results_df = pd.DataFrame()
    for samp in OverSamp_list:
        X_train_resampled, y_train_resampled = samp.fit_resample(X_train, y_train)
        print('Fitting Model ... in', samp)
        model.fit(X_train_resampled, y_train_resampled)
        y_pred = model.predict(X_test)
        score_df = pd.DataFrame({'model': str(model[-1]).split('(')[0],
                             'f1': f1_score(y_test, y_pred),
                             'auc': roc_auc_score(y_test, y_pred)},
                             index=[str(samp).split('(')[0]])

        results_df = pd.concat([results_df, score_df])
    result_df = pd.concat([results_df, result_df], axis=0)

result_df = result_df.sort_values(by='f1', ascending=False)
result_df.round(4)
```

So far, the best performing model is the Random Forest Classifier  using the Smote oversampling techniques to resampling the training sets, with an f1 score of 0.19 and auc score of 0.59.
